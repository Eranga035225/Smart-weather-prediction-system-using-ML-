```python
import pandas as pd
```

```python
# Step 1: Load CSV files
location_df = pd.read_csv('../../data/locationData.csv')
weather_df = pd.read_csv('../../data/weatherData.csv')
```

```python
location_df.head()
```

```python
weather_df.head()
```

```python
# Merge on 'location_id'
merged_df = pd.merge(weather_df, location_df, on='location_id', how='left')
```

```python
# Optional: Check merged rows count matches weather_df rows
print(f"Weather rows: {len(weather_df)}, Merged rows: {len(merged_df)}")
```

```python
# Save to a new CSV if needed
merged_df.to_csv('merged_weather_location.csv', index=False)
```
